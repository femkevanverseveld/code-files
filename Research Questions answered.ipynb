```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# RQ2: To what extent can various ML models accurately classify the design attributes of the virtual character, encompassing Expressiveness, 
# after being trained on Mouth, Left Eye, Right Eye, Left Eyebrow, and Right Eyebrow, using dimensionality reduction techniques?

# Data organized into structured format
data_Expressiveness_F = {
    'Model': ['Logistic Regression', 'Random Forest', 'Logistic Regression', 'Random Forest', 'Logistic Regression', 'Random Forest', 'Logistic Regression', 'Random Forest', 'Logistic Regression', 'Random Forest', 'Logistic Regression', 'Random Forest'],
    'Data Type': ['Base', 'Base', 'PCA', 'PCA', 'KPCA', 'KPCA', 'Base', 'Base', 'PCA', 'PCA', 'KPCA', 'KPCA'],
    'Average CV AU-ROC': [0.5972, 0.6833, 0.5639, 0.6028, 0.5722, 0.6361, 0.5972, 0.6833, 0.5639, 0.6028, 0.5722, 0.6361],
    'Test AU-ROC': [0.5312, 0.5156, 0.5000, 0.5234, 0.4688, 0.5625, 0.5312, 0.5156, 0.5000, 0.5234, 0.4688, 0.5625],
    'Average F1-Score': [0.44, 0.55, 0.44, 0.60, 0.44, 0.49, 0.44, 0.55, 0.44, 0.60, 0.44, 0.49]
}

data_Expressiveness_F = pd.DataFrame(data_Expressiveness_F)

# Get colors from seaborn palette
sns_colors = sns.color_palette("viridis", data_Expressiveness_F['Model'].nunique())
colors_dict = dict(zip(data_Expressiveness_F['Model'].unique(), sns_colors))

# Define the palette
palette = {"Logistic Regression": colors_dict['Logistic Regression'], "Random Forest": colors_dict['Random Forest']}

# Create plots
metrics = ['Average CV AU-ROC', 'Test AU-ROC', 'Average F1-Score']
for metric in metrics:
    plt.figure(figsize=(15,6))
    plot = sns.barplot(x='Data Type', y=metric, hue='Model', data=data_Expressiveness_F, palette=palette)
    plt.title(metric + ' for Logistic Regression and Random Forest on Facial States in Classifying Expressiveness')
    
    # Display numbers on the bars
    for p in plot.patches:
        plot.annotate(format(p.get_height(), '.2f'), 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha = 'center', va = 'center', 
                      xytext = (0, 10), 
                      textcoords = 'offset points')
        
    # Design aspects
    plot.spines['top'].set_visible(False)
    plot.spines['right'].set_visible(False)
    plot.grid(axis='y', linestyle='--', alpha=0.7)

    # Position the legend on the upper right
    plt.legend(loc='right')
    plt.tight_layout()

plt.show()
```

```python
# RQ2: To what extent can various ML models accurately classify the design attributes of the virtual character, encompassing Artificiality and Expressiveness, 
# after being trained on Mouth, Left Eye, Right Eye, Left Eyebrow, and Right Eyebrow, using dimensionality reduction techniques?

# Reshape dataframes to long format
data_Expressiveness_F_long = data_Expressiveness_F.melt(id_vars=['Model', 'Data Type'], 
                                                        value_vars=['Average CV AU-ROC', 'Test AU-ROC', 'Average F1-Score'],
                                                        var_name='Metric', value_name='Value')
data_Expressiveness_F_long['Dataset'] = 'Expressiveness_F'

data_Artificiality_F_long = data_Artificiality_F.melt(id_vars=['Model', 'Data Type'], 
                                                      value_vars=['Average CV AU-ROC', 'Test AU-ROC', 'Average F1-Score'],
                                                      var_name='Metric', value_name='Value')
data_Artificiality_F_long['Dataset'] = 'Artificiality_F'

# Concatenate reshaped dataframes
combined_data = pd.concat([data_Expressiveness_F_long, data_Artificiality_F_long])

# Create a combined hue
combined_data['Hue'] = combined_data['Dataset'] + ' - ' + combined_data['Model']

# Plot
plt.figure(figsize=(15, 8))
barplot = sns.barplot(x='Metric', y='Value', hue='Hue', data=combined_data, ci=None, palette='viridis', dodge=True)

# Add the values on top of the bars
for i, bar in enumerate(barplot.patches):
    # Annotate with the bar's value
    barplot.annotate(format(bar.get_height(), '.2f'), 
                     (bar.get_x() + bar.get_width() / 2., bar.get_height()), 
                     ha = 'center', va = 'center', 
                     xytext = (0, 9), 
                     textcoords = 'offset points')

plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
