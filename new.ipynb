```python
# Remove 'FIND_FAILED', 'FIT_FAILED', and 'NA' values
values_to_remove = ["FIND_FAILED", "FIT_FAILED"]

# Create new list to hold the processed dataframes
processed_dataframes = []

for df in data_frames:
    # Drop NA values
    df.dropna(inplace=True)
    
    # Remove rows containing "FIND_FAILED" or "FIT_FAILED" in any column
    df = df[~df.isin(values_to_remove).any(axis=1)]
    
    processed_dataframes.append(df)

# Convert list back to a tuple
data_frames = tuple(processed_dataframes)
```

```python
# Create list of columns to drop
columns_to_drop = ['Condition', 'Arousal', 'Valence', 'Video Time']

# List comprehension to drop the columns for each dataframe in the tuple
data_frames = [df.drop(columns=columns_to_drop, errors='ignore') for df in data_frames]
```

```python
# Concatenate data frames into a single data frame
dataset = pd.concat(data_frames)

# Display data frame
dataset
```

```python
# One-Hot Encoding
df_encoded = pd.get_dummies(dataset, columns=['Mouth', 'Left Eye', 'Right Eye', 'Left Eyebrow', 'Right Eyebrow'])

expected_columns = [
    'Mouth_Open', 'Mouth_Closed',
    'Left Eye_Open', 'Left Eye_Closed',
    'Right Eye_Open', 'Right Eye_Closed',
    'Left Eyebrow_Neutral', 'Left Eyebrow_Raised', 'Left Eyebrow_Lowered',
    'Right Eyebrow_Neutral', 'Right Eyebrow_Raised', 'Right Eyebrow_Lowered'
]

for col in expected_columns:
    if col not in df_encoded.columns:
        df_encoded[col] = 0

# Create aggregation functions dynamically for expected_columns and additional columns
additional_cols = ['Neutral', 'Happy', 'Sad', 'Angry', 'Surprised', 'Scared', 'Disgusted']
agg_funcs = {col: ['mean', 'std', 'skew', 'median', lambda x: x.max() - x.min()] for col in (expected_columns + additional_cols)}

# Group by 'ID' and aggregate statistics
df_aggregated = df_encoded.groupby('ID').agg(agg_funcs)

# Custom column naming
df_aggregated.columns = ['_'.join(col).strip().replace('<lambda_0>', 'range') for col in df_aggregated.columns.values]

# Add back 'Artificiality' and 'Expressiveness'
artificiality_expressiveness = dataset[['ID', 'Artificiality', 'Expressiveness']].drop_duplicates().set_index('ID')
df_aggregated = df_aggregated.join(artificiality_expressiveness)

# Reset the index to make 'ID' a column
df_aggregated = df_aggregated.reset_index()

# Display data frame
df_aggregated
```

```python
# Filter out rows where 'Artificiality' or 'Expressiveness' is 3
df_aggregated = df_aggregated[(df_aggregated['Artificiality'] != 3) & (df_aggregated['Expressiveness'] != 3)]
```

```python
# Exclude 'ID' from df_aggregated
data = df_aggregated.iloc[:, 1:]
```

```python
# Display final data frame
data
```

```python
# Feature matrix and 'Expressiveness' target
X = data.drop(['Artificiality', 'Expressiveness'], axis=1)
y_expressiveness = data['Expressiveness']

# Split the data into training and test sets before any scaling to prevent data leakage
X_train, X_test, y_train, y_test = train_test_split(
    X, y_expressiveness, test_size=0.2, random_state=42, stratify=y_expressiveness)

# Standardize the features based on the training set
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Fit the LDA model on the scaled training data
lda = LDA(n_components=1)
X_train_lda = lda.fit_transform(X_train_scaled, y_train)

# Get the absolute values of the loadings for each feature
lda_coefs = np.abs(lda.scalings_).ravel()

# Calculate the sorted cumulative importance of the features
sorted_idx = np.argsort(lda_coefs)[::-1]
sorted_cumulative_importance = np.cumsum(lda_coefs[sorted_idx]) / np.sum(lda_coefs)

# Find the number of features needed to reach 95% cumulative importance
num_features_95_percent = np.where(sorted_cumulative_importance >= 0.95)[0][0] + 1

# Select the features that contribute to 95% of cumulative importance
selected_features_indices = sorted_idx[:num_features_95_percent]
selected_features = X_train.columns[selected_features_indices]
```

```python
# Define the training and test data for the top features selected by LDA
X_train_top = X_train_scaled[:, selected_features_indices]
X_test_top = X_test_scaled[:, selected_features_indices]

# Logistic Regression on the top features
param_grid_lr = {'C': [0.001, 0.01, 0.1, 1]}
grid_search_lr = GridSearchCV(LogisticRegression(max_iter=1000), param_grid=param_grid_lr, cv=10, scoring='f1', n_jobs=-1)
grid_search_lr.fit(X_train_top, y_train)

# Best Logistic Regression model
best_model_lr = grid_search_lr.best_estimator_

# Get the coefficients from the Logistic Regression model
lr_coefs = best_model_lr.coef_.ravel()

# Match the coefficients to the selected features
feature_importance_lr = pd.Series(lr_coefs, index=selected_features)

# Sort the features by the absolute value of the coefficients
sorted_features = feature_importance_lr.abs().sort_values(ascending=False).index

# Reorder the coefficients
sorted_coefs = feature_importance_lr.loc[sorted_features]

# Plot the feature importances
plt.figure(figsize=(15, 6))
sorted_coefs.plot(kind='bar')
plt.title('Logistic Regression Feature Importance')
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.xticks(rotation=90)  # Rotate the feature names for better readability
plt.tight_layout()
plt.show()

# Evaluate the Logistic Regression model and print classification reports
y_train_pred_lr = best_model_lr.predict(X_train_top)
y_test_pred_lr = best_model_lr.predict(X_test_top)

print('Classification Report for Training Data:')
print(classification_report(y_train, y_train_pred_lr))

print('Classification Report for Test Data:')
print(classification_report(y_test, y_test_pred_lr))

print(f"Training F1-score LR: {f1_score(y_train, y_train_pred_lr)}")
print(f"Test F1-score LR: {f1_score(y_test, y_test_pred_lr)}")
```

```python
# Define the training and test data for the top features selected by LDA
X_train_top = X_train_scaled[:, selected_features_indices]
X_test_top = X_test_scaled[:, selected_features_indices]

# Support Vector Machine on the top features with a linear kernel
param_grid_svm_linear = {'C': [0.001, 0.01, 0.1, 1]}
grid_search_svm_linear = GridSearchCV(SVC(kernel='linear'), param_grid=param_grid_svm_linear, cv=10, scoring='f1', n_jobs=-1)
grid_search_svm_linear.fit(X_train_top, y_train)

# Best SVM model with linear kernel
best_model_svm_linear = grid_search_svm_linear.best_estimator_

# Get the coefficients from the linear SVM model
svm_coefs = best_model_svm_linear.coef_.ravel()

# Match the coefficients to the selected features
feature_importance_svm = pd.Series(svm_coefs, index=selected_features)

# Get the sorted feature importances in descending order
sorted_idx = np.argsort(feature_importance_svm)[::-1]

# Plot the sorted feature importances
plt.figure(figsize=(15, 6))
feature_importance_svm_sorted = feature_importance_svm.iloc[sorted_idx]
feature_importance_svm_sorted.plot(kind='bar')
plt.title('SVM (Linear Kernel) Feature Importance')
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.xticks(rotation=90)  # Rotate the feature names for better readability
plt.tight_layout()
plt.show()

# Evaluate the SVM model with a linear kernel and print classification reports
y_train_pred_svm_linear = best_model_svm_linear.predict(X_train_top)
y_test_pred_svm_linear = best_model_svm_linear.predict(X_test_top)

print('Classification Report for Training Data:')
print(classification_report(y_train, y_train_pred_svm_linear))

print('Classification Report for Test Data:')
print(classification_report(y_test, y_test_pred_svm_linear))

print(f"Training F1-score SVM (Linear Kernel): {f1_score(y_train, y_train_pred_svm_linear)}")
print(f"Test F1-score SVM (Linear Kernel): {f1_score(y_test, y_test_pred_svm_linear)}")
```
