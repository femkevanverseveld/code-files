```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import plot_confusion_matrix, confusion_matrix
import itertools
import seaborn as sns
```

```python
# Import 'data2.csv'
dataset = pd.read_csv('data2.csv')
```

```python
# Display data frame
dataset
```

```python
# VIF test for multicollinearity
X_Expressiveness = dataset.drop(['Expressiveness','Artificiality'], axis=1)  
y_Expressiveness = dataset['Expressiveness'] 

# VIF for each predictor variable
X_Expressiveness = sm.add_constant(X_Expressiveness) 
vif = pd.Series([variance_inflation_factor(X_Expressiveness.values, i) for i in range(X_Expressiveness.shape[1])],
                index=X_Expressiveness.columns)

# Variables above the threshold
vif_threshold = 10  
multicollinear_vars = vif[vif > vif_threshold]  
num_multicollinear_vars = len(multicollinear_vars)
num_non_multicollinear_vars = len(vif) - num_multicollinear_vars

# Display VIF above/below threshold and multicollinear variables
print("Number of variables with VIF above threshold ({}): {}".format(vif_threshold, num_multicollinear_vars))
print("Number of variables with VIF below threshold ({}): {}".format(vif_threshold, num_non_multicollinear_vars))
print("Multicollinear variables:", multicollinear_vars.index.tolist())
```

```python
# Preparing the data for modeling Expressiveness Base
# Extract features by dropping target columns from the dataset
X_Expressiveness_base = dataset.drop(['Expressiveness', 'Artificiality'], axis=1)

# Extract target variable 'Expressiveness' from the dataset
y_Expressiveness_base = dataset['Expressiveness']

# Split the data into training and testing sets (80% training, 20% testing)
# The split is stratified by the target variable 'Expressiveness' to maintain the distribution in both sets
X_train_Expressiveness_base, X_test_Expressiveness_base, y_train_Expressiveness_base, y_test_Expressiveness_base = train_test_split(X_Expressiveness_base, y_Expressiveness_base, test_size=0.2, random_state=42, stratify=y_Expressiveness_base)
```

```python
# Evaluation function to print model metrics
def evaluate_and_print_metrics(model_name, grid_search, X_test, y_test):
    # Predict class labels for the test set using best estimator from grid search
    predictions = grid_search.best_estimator_.predict(X_test)
    
    # Calculate F1-score for the model on test data
    f1_value = f1_score(y_test, predictions, average='binary')
    
    # Print model name and its metrics
    print(f"{model_name} Metrics:")
    print(f"Average CV F1-Score: {grid_search.best_score_:.4f}")
    print(f"Test F1-Score: {f1_value:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, predictions))
    print("\n")

# Define hyperparameter grid for Logistic Regression
param_grid_lr = {'C': [0.001, 0.01, 0.1]}

# Apply GridSearchCV for hyperparameter tuning on Logistic Regression model
grid_search_lr_Expressiveness_base = GridSearchCV(LogisticRegression(max_iter=100, penalty='l2'), param_grid=param_grid_lr, cv=10, scoring='f1', n_jobs=-1)
grid_search_lr_Expressiveness_base.fit(X_train_Expressiveness_base, y_train_Expressiveness_base)

# Evaluate and print metrics for Logistic Regression (Expressiveness Base) model
evaluate_and_print_metrics("Logistic Regression (Expressiveness Base)", grid_search_lr_Expressiveness_base, X_test_Expressiveness_base, y_test_Expressiveness_base)

# Create a Random Forest Classifier instance
rf_Expressiveness_base = RandomForestClassifier()

# Define hyperparameter grid for Random Forest
param_grid_rf_Expressiveness_base = {
    'n_estimators': [10, 50, 100, 150],
    'max_depth': [None, 5, 10, 15, 20],
    'min_samples_split': [2, 3, 5, 7, 10],
    'min_samples_leaf': [1, 2, 3, 4],
    'max_features': ['auto', 'sqrt', 'log2', None]
}

# Apply GridSearchCV for hyperparameter tuning on Random Forest model
grid_search_rf_Expressiveness_base = GridSearchCV(rf_Expressiveness_base, param_grid=param_grid_rf_Expressiveness_base, cv=10, scoring='f1', n_jobs=-1)
grid_search_rf_Expressiveness_base.fit(X_train_Expressiveness_base, y_train_Expressiveness_base)

# Evaluate and print metrics for Random Forest (Expressiveness Base) model
evaluate_and_print_metrics("Random Forest (Expressiveness Base)", grid_search_rf_Expressiveness_base, X_test_Expressiveness_base, y_test_Expressiveness_base)
```

```python
# Prepare data for PCA analysis targeting Expressiveness
# Remove target variables to create features set
X_Expressiveness_pca = dataset.drop(['Expressiveness', 'Artificiality'], axis=1)
# Extract Expressiveness as the target variable
y_Expressiveness_pca = dataset['Expressiveness']

# Initialize PCA and fit-transform the features data
Expressiveness_pca = PCA()
X_Expressiveness_pca = Expressiveness_pca.fit_transform(X_Expressiveness_pca)

# Calculate the explained variance for each principal component
explained_variance_ratio_Expressiveness_pca = Expressiveness_pca.explained_variance_ratio_
# Calculate the cumulative explained variance across all components
cumulative_variance_ratio_Expressiveness_pca = np.cumsum(explained_variance_ratio_Expressiveness_pca)

# Set a 95% threshold for the desired cumulative variance
threshold_pca = 0.95
# Find the minimum number of principal components needed to reach or exceed the threshold
num_components_Expressiveness_pca = np.argmax(cumulative_variance_ratio_Expressiveness_pca >= threshold_pca) + 1

# Plotting the scree plot to visualize cumulative explained variance across principal components
plt.figure(figsize=(25, 8))
plt.plot(range(1, len(cumulative_variance_ratio_Expressiveness_pca) + 1), cumulative_variance_ratio_Expressiveness_pca, 'bo-', label='Cumulative Explained Variance')
plt.axhline(y=threshold_pca, color='g', linestyle='--', label=f'{threshold_pca*100:.1f}% Cumulative Variance Threshold')
plt.title('Scree Plot for Expressiveness PCA')
plt.xlabel('Principal Component Number')
plt.ylabel('Value')
plt.xticks(range(1, len(explained_variance_ratio_Expressiveness_pca) + 1))
plt.legend()
plt.show()

# Print the computed number of principal components required to meet the variance threshold
print(f"Number of Principal Components for Expressiveness with PCA at {threshold_pca*100:.1f}% Cumulative Variance: {num_components_Expressiveness_pca}")
```

```python
# Reduce the dimensionality of the data using the selected number of principal components for Expressiveness
X_reduced_Expressiveness_pca = X_Expressiveness_pca[:, :num_components_Expressiveness_pca]

# Split the reduced data into training and testing sets (80% training, 20% testing)
# The split is stratified by the target variable 'Expressiveness' to maintain the distribution in both sets
X_train_Expressiveness_pca, X_test_Expressiveness_pca, y_train_Expressiveness_pca, y_test_Expressiveness_pca = train_test_split(X_reduced_Expressiveness_pca, y_Expressiveness_pca, test_size=0.2, random_state=42, stratify=y_Expressiveness_pca)
```

```python
# Define an evaluation function to assess model performance
def evaluate_and_print_metrics(model_name, grid_search, X_test, y_test):
    # Predict class labels for the test set using best estimator from grid search
    predictions = grid_search.best_estimator_.predict(X_test)
    
    # Calculate F1-score for the model on test data
    f1_value = f1_score(y_test, predictions, average='binary')
    
    # Print model name and its metrics
    print(f"{model_name} Metrics:")
    print(f"Average CV F1-Score: {grid_search.best_score_:.4f}")
    print(f"Test F1-Score: {f1_value:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, predictions))
    print("\n")

# Define hyperparameter grid for Logistic Regression
param_grid_lr = {'C': [0.001, 0.01, 0.1]}

# Apply GridSearchCV for hyperparameter tuning on Logistic Regression model
grid_search_lr_Expressiveness_pca = GridSearchCV(LogisticRegression(max_iter=100, penalty='l2'), param_grid=param_grid_lr, cv=10, scoring='f1', n_jobs=-1)
grid_search_lr_Expressiveness_pca.fit(X_train_Expressiveness_pca, y_train_Expressiveness_pca)

# Evaluate and print metrics for Logistic Regression (Expressiveness PCA) model
evaluate_and_print_metrics("Logistic Regression (Expressiveness PCA)", grid_search_lr_Expressiveness_pca, X_test_Expressiveness_pca, y_test_Expressiveness_pca)

# Create a Random Forest Classifier instance
rf_Expressiveness_pca = RandomForestClassifier()

# Define hyperparameter grid for Random Forest
param_grid_rf_Expressiveness_pca = {
    'n_estimators': [10, 50, 100, 150],
    'max_depth': [None, 5, 10, 15, 20],
    'min_samples_split': [2, 3, 5, 7, 10],
    'min_samples_leaf': [1, 2, 3, 4],
    'max_features': ['auto', 'sqrt', 'log2', None]
}

# Apply GridSearchCV for hyperparameter tuning on Random Forest model
grid_search_rf_Expressiveness_pca = GridSearchCV(rf_Expressiveness_pca, param_grid=param_grid_rf_Expressiveness_pca, cv=10, scoring='f1', n_jobs=-1)
grid_search_rf_Expressiveness_pca.fit(X_train_Expressiveness_pca, y_train_Expressiveness_pca)

# Evaluate and print metrics for Random Forest (Expressiveness PCA) model
evaluate_and_print_metrics("Random Forest (Expressiveness PCA)", grid_search_rf_Expressiveness_pca, X_test_Expressiveness_pca, y_test_Expressiveness_pca)
```

```python
# Extracting features and target variable for Kernel PCA (KPCA)
# Removing the 'Expressiveness' and 'Artificiality' columns to get the feature matrix
X_Expressiveness_kpca = dataset.drop(['Expressiveness', 'Artificiality'], axis=1)
# Getting the target variable 'Expressiveness'
y_Expressiveness_kpca = dataset['Expressiveness']

# Define a list of candidate gamma values for the RBF kernel in KPCA
gammas = [0.001, 0.01, 0.1, 1, 10]

# Initialize variables to find the optimal gamma
best_gamma = None
# Initially set the lowest number of components to the total number of original features
lowest_num_components = len(X_Expressiveness_kpca.columns)

# Iterate over each candidate gamma value to find the optimal one
for gamma in gammas:
    # Apply KPCA with the RBF kernel and current gamma value
    Expressiveness_kpca = KernelPCA(kernel="rbf", gamma=gamma, n_components=None)
    X_E_transformed = Expressiveness_kpca.fit_transform(X_Expressiveness_kpca)

    # Compute the explained variance ratio for each principal component
    explained_variance_ratio = np.var(X_E_transformed, axis=0) / np.sum(np.var(X_E_transformed, axis=0))
    # Calculate the cumulative explained variance ratio
    cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

    # Determine the number of components where 95% (or a similar threshold) of the variance is retained
    threshold_kpca = 0.95
    num_components = np.argmax(cumulative_variance_ratio >= threshold_kpca) + 1

    # Update the best gamma if the current one results in fewer principal components
    if num_components < lowest_num_components:
        lowest_num_components = num_components
        best_gamma = gamma

# Display the best gamma value and the number of components for that gamma
print(f"Best gamma value: {best_gamma}")
print(f"Number of Principal Components with gamma={best_gamma} at {threshold_kpca*100:.1f}% Cumulative Variance: {lowest_num_components}")

# Use the best gamma value to transform the data for visualization
Expressiveness_kpca_best = KernelPCA(kernel="rbf", gamma=best_gamma, n_components=None)
X_Expressiveness_kpca_best = Expressiveness_kpca_best.fit_transform(X_Expressiveness_kpca)

# Compute the explained variance ratio for the transformed data using the best gamma
explained_variance_ratio_best = np.var(X_Expressiveness_kpca_best, axis=0) / np.sum(np.var(X_Expressiveness_kpca_best, axis=0))
cumulative_variance_ratio_best = np.cumsum(explained_variance_ratio_best)

# Visualization: scree plot to show the cumulative explained variance for each principal component
plt.figure(figsize=(25, 8))
plt.plot(range(1, len(cumulative_variance_ratio_best) + 1), cumulative_variance_ratio_best, 'bo-', label='Cumulative Explained Variance with KPCA')
plt.axhline(y=threshold_kpca, color='g', linestyle='--', label=f'{threshold_kpca*100:.1f}% Cumulative Variance Threshold with KPCA')
plt.title(f'Scree Plot for Expressiveness KPCA (gamma={best_gamma})')
plt.xlabel('Principal Component Number')
plt.ylabel('Value')
plt.xticks(range(1, len(explained_variance_ratio_best) + 1))
plt.legend()
plt.show()
```

```python
# Reduce dimensionality using principal components for Expressiveness from KPCA
X_reduced_Expressiveness_kpca = X_Expressiveness_kpca_best[:, :lowest_num_components]

# Stratified split of the reduced data into 80% training and 20% testing sets
X_train_Expressiveness_kpca, X_test_Expressiveness_kpca, y_train_Expressiveness_kpca, y_test_Expressiveness_kpca = train_test_split(X_reduced_Expressiveness_kpca, y_Expressiveness_kpca, test_size=0.2, random_state=42, stratify=y_Expressiveness_kpca)
```

```python
# Define an evaluation function to assess model performance
def evaluate_and_print_metrics(model_name, grid_search, X_test, y_test):
    # Predict class labels for the test set using best estimator from grid search
    predictions = grid_search.best_estimator_.predict(X_test)
    
    # Calculate F1-score for the model on test data
    f1_value = f1_score(y_test, predictions, average='binary') 
    
    # Print model name and its metrics
    print(f"{model_name} Metrics:")
    print(f"Average CV F1-Score: {grid_search.best_score_:.4f}")
    print(f"Test F1-Score: {f1_value:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, predictions))
    print("\n")

# Define hyperparameter grid for Logistic Regression
param_grid_lr = {'C': [0.001, 0.01, 0.1]}

# Apply GridSearchCV for hyperparameter tuning on Logistic Regression model
grid_search_lr_Expressiveness_kpca = GridSearchCV(LogisticRegression(max_iter=100, penalty='l2'), param_grid=param_grid_lr, cv=10, scoring='f1', n_jobs=-1)
grid_search_lr_Expressiveness_kpca.fit(X_train_Expressiveness_kpca, y_train_Expressiveness_kpca)

# Evaluate and print metrics for Logistic Regression (Expressiveness KPCA) model
evaluate_and_print_metrics("Logistic Regression (Expressiveness KPCA)", grid_search_lr_Expressiveness_kpca, X_test_Expressiveness_kpca, y_test_Expressiveness_kpca)

# Create a Random Forest Classifier instance
rf_Expressiveness_kpca = RandomForestClassifier()

# Define hyperparameter grid for Random Forest
param_grid_rf_Expressiveness_kpca = {
    'n_estimators': [10, 50, 100, 150],
    'max_depth': [None, 5, 10, 15, 20],
    'min_samples_split': [2, 3, 5, 7, 10],
    'min_samples_leaf': [1, 2, 3, 4],
    'max_features': ['auto', 'sqrt', 'log2', None]
}

# Apply GridSearchCV for hyperparameter tuning on Random Forest model
grid_search_rf_Expressiveness_kpca = GridSearchCV(rf_Expressiveness_kpca, param_grid=param_grid_rf_Expressiveness_kpca, cv=10, scoring='f1', n_jobs=-1)
grid_search_rf_Expressiveness_kpca.fit(X_train_Expressiveness_kpca, y_train_Expressiveness_kpca)

# Evaluate and print metrics for Random Forest (Expressiveness KPCA) model
evaluate_and_print_metrics("Random Forest (Expressiveness KPCA)", grid_search_rf_Expressiveness_kpca, X_test_Expressiveness_kpca, y_test_Expressiveness_kpca)
```

```python
def evaluate_and_return_metrics(grid_search, X_test, y_test):
    # Compute average CV F1-Score and Test F1-Score
    avg_cv_f1 = grid_search.best_score_
    predictions = grid_search.best_estimator_.predict(X_test)
    test_f1 = f1_score(y_test, predictions, average='binary')
    return avg_cv_f1, test_f1

# Evaluate models and store results
results = {
    "LR (Base)": evaluate_and_return_metrics(grid_search_lr_Expressiveness_base, X_test_Expressiveness_base, y_test_Expressiveness_base),
    "RF (Base)": evaluate_and_return_metrics(grid_search_rf_Expressiveness_base, X_test_Expressiveness_base, y_test_Expressiveness_base),
    "LR (PCA)": evaluate_and_return_metrics(grid_search_lr_Expressiveness_pca, X_test_Expressiveness_pca, y_test_Expressiveness_pca),
    "RF (PCA)": evaluate_and_return_metrics(grid_search_rf_Expressiveness_pca, X_test_Expressiveness_pca, y_test_Expressiveness_pca),
    "LR (KPCA)": evaluate_and_return_metrics(grid_search_lr_Expressiveness_kpca, X_test_Expressiveness_kpca, y_test_Expressiveness_kpca),
    "RF (KPCA)": evaluate_and_return_metrics(grid_search_rf_Expressiveness_kpca, X_test_Expressiveness_kpca, y_test_Expressiveness_kpca)
}

# Extract names and scores for plotting
names = list(results.keys())
avg_cv_values = [item[0] for item in results.values()]
test_values = [item[1] for item in results.values()]

# Define colors and hatch patterns for plotting
colors = {
    "LR": "lightblue",
    "RF": "darkblue"
}
hatch_patterns = {
    "Avg CV F1-Score": '',
    "Test F1-Score": 'xxxx'
}

# Initialize plot settings
bar_width = 0.35
index = np.arange(len(names))
plt.figure(figsize=(15, 6))

# Plot bars for each model type
for i, (name, avg_cv, test) in enumerate(zip(names, avg_cv_values, test_values)):
    model_type = name.split(" ")[0]
    # Plot bars with corresponding color and hatch pattern
    plt.bar(i - bar_width/2, avg_cv, color=colors[model_type], width=bar_width, label=f"CV F1-Score" if i == 0 else "")
    plt.text(i - bar_width/2, avg_cv + 0.01, f"{avg_cv:.2f}", ha='center', fontsize=10)
    
    plt.bar(i + bar_width/2, test, color=colors[model_type], hatch=hatch_patterns["Test F1-Score"], width=bar_width, label=f"Test F1-Score" if i == 0 else "")
    plt.text(i + bar_width/2, test + 0.01, f"{test:.2f}", ha='center', fontsize=10)

# Configure plot
plt.ylabel('F1-Score')
plt.title('Model Performance Comparison')
plt.xticks(index, names)
plt.legend(loc='upper right')
plt.ylim([0, 1.1])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
